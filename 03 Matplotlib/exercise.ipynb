```python
# create a python dictionary of 3 your favorite foods with
# the keys of the dictionary should be the food name and the values
favorite_food_price = {"Almond": 10, "Blueberry": 5, "Banana": 6}

# create a bar graph where the x-axis is the keys of the dictionary
# and the y-axis is the values of the dictionary
fig, ax = plt.subplots()
ax.bar(favorite_food_price.keys(), favorite_food_price.values())

# add a tittle, xlabel and ylabel to the plot
ax.set(title="Ikmal's Favorite Food", xlabel="Food", ylabel="Price ($)")
```

```python
# make the same plot as above, except this time make the bars go horizontal
fig, ax = plt.subplots()
ax.barh(list(favorite_food_price.keys()), list(favorite_food_price.values()))
ax.set(title="Ikmal's Favorite Food", xlabel="Price ($)", ylabel="Food")
```

```python
car_sales = pd.read_csv("../01 Pandas Data Analysis/car-sales.csv")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace("[$\\,\\.]", "", regex=True)
car_sales["Price"]
```

```python
# remove the symbol, the final 2 numbers from the 'Price' column and convert it to numbers
car_sales["Price"] = car_sales["Price"].str.replace("[\\$\\,\\.]", "", regex=True)
car_sales["Price"] = car_sales["Price"].str[:2]

# add a column called 'Total Sales to car_sales which cumulatively adds the 'price' column
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()

# add a column called 'Sale Date' which list a series of successive dates starting from today
car_sales["Sales Date"] = pd.date_range("23/06/2024", periods=len(car_sales))

# view the car_sales DataFrame
car_sales
```

```python
# use the plot() function to plot the "Sale Date" column versus the 'Total Sales' column
car_sales.plot(x="Sales Date", y="Total Sales")
```

```python
heart_disease = pd.read_csv("./heart-disease.csv")
heart_disease.head(10)
```

```python
# create a histogram of the 'age' column with 50 bins
heart_disease["age"].plot.hist(bins=50)
```

```python
# call plot.hist() on the heart_disease DataFrame and toggle the 'subplots' parameter to true
heart_disease.plot.hist(subplots=True)
```

```python
# call the same line of code from above except change the 'figsize' parameter to be (10, 3)
heart_disease.plot.hist(figsize=(10, 30), subplots=True)
```

```python
# replicated the above plot in whichever way you see fit

# note: the method bellow is only one way of doing it, your might be slightly different

# create DataFrame with patients over 50 y.o
over_50 = heart_disease[heart_disease["age"] > 50]

# create a plot
fig, ax = plt.subplots(figsize=(10, 6))

# plot the dat
scatter = ax.scatter(over_50["age"], over_50["chol"], c=over_50["target"])

# customize the plot
ax.set(title="Heart Disease and Cholesterol levels", xlabel="Age", ylabel="Cholesterol")
ax.legend(*scatter.legend_elements(), title="Target")

# add a mealine
ax.axhline(over_50["chol"].mean(), linestyle="--")
```
